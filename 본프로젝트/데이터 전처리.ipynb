```python
import numpy as np
import pandas as pd

import re
```

```python
df = pd.read_csv('data/total.csv',index_col = 0)
df
```

```python
def preprocess(x):
    condition = "\[.*\]|\s-\s.*" 
    x = re.sub(condition, '',x)
    return x
```

```python
df['title']=df['title'].apply(preprocess)
df.to_csv('data/total.csv',encoding='utf-8-sig')
```

```python
df=df.reset_index(drop=True)
df
```

```python
df['len']=df['title'].apply(len)
```

```python
index=df[df['len']<=7].index
df.drop(index,inplace=True)
```

```python
df=df.reset_index(drop=True)
df
```

```python
df.drop('len',axis='columns',inplace=True)
df
```

```python
df.to_csv('data/total.csv',encoding='utf-8-sig')
```
