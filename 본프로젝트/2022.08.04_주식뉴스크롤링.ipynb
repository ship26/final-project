```python
from bs4 import BeautifulSoup
import requests

titles = []
for page_num in tqdm(range(1,11)):
    code='097950'
    url_front = f"https://finance.naver.com/item/news_news.nhn?code={code}&page="
    url_last = "&sm=title_entity_id.basic&clusterId="
    url = url_front + str(page_num) + url_last
    print(url)
    html = requests.get(url).content
    bs = BeautifulSoup(html,'html.parser')
    
    all_tr = bs.findAll('td', {'class' : 'title'})
    
    time.sleep(1)

    
    for tr in all_tr:
        titles.append(tr.find('a').get_text())
```
