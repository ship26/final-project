## 2022.4.2 한국경제 타이틀크롤링 27면

```python
response=requests.get('https://media.naver.com/press/015/newspaper',
             params={'date':'20220402'}
            )
```

```python
soup=BeautifulSoup(response.text)
soup
```

```python

면=soup.select('div.newspaper_inner')
```

```python
타이틀제목들=[]
for i in range(len(면)):
    i면타이틀=[]
    타이틀제목=soup.select('div.newspaper_inner')[i]
    타이틀제목s=타이틀제목.find_all('strong')
    for j in 타이틀제목s:
        i면타이틀.append(j.text)
    타이틀제목들.append(i면타이틀)
타이틀제목들
```

## 한국경제 2022.4월동안의 뉴스타이틀을 면순서대로 자동크롤링해보기

```python
한국경제_타이틀_4월=[]
for i in tqdm(range(1,31)):
    하루타이틀=[]
    if i <10:
        response=requests.get('https://media.naver.com/press/015/newspaper',
                     params={'date':'2022040{}'.format(i)}
                    )
    else:
        response=requests.get('https://media.naver.com/press/015/newspaper',
                     params={'date':'202204{}'.format(i)}
                    )
        
        
    soup=BeautifulSoup(response.text)
    면=soup.select('div.newspaper_inner')
    
    if len(면)<1:
        print(f'{i}일자는 기사가 없습니다 데헷')
        print('-'*25)
        continue
    else:
        print(f'{i}일자 기사 크롤링합니다.')
        print('-'*25)
        for i in range(len(면)):
            i면타이틀=[]
            타이틀제목=soup.select('div.newspaper_inner')[i]
            타이틀제목s=타이틀제목.find_all('strong')
            for j in 타이틀제목s:
                i면타이틀.append(j.text)
            하루타이틀.append(i면타이틀)
        
        한국경제_타이틀_4월.append(하루타이틀)
```
