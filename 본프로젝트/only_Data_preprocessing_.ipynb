```python
def clean_text(str):
  text_rmv = re.sub('[-=+,#/\?:^.@*\"※ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', ' ', str)
  return text_rmv
```

```python
clean_text(df["내용"][1])
```

```python
df['clean_text'] = df['내용'].apply(clean_text)
```
