```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd
import numpy as np
```

```python
# 언론사 리스트
press_list = {'한국경제':'015', '매일경제':'009', '조선일보':'023', '동아일보':'020', '한겨례신문':'028', '경향신문':'032'}

# 크롤링 할 대상 입력
to_crawl_press_this_time = input("크롤링 할 언론사 : ")
presscode = press_list[to_crawl_press_this_time]


# 빈 DF 생성
df = pd.DataFrame(columns=["press","title","date"])

request_headers = { 
'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36' } 
# 일년 반복문 
for month in range(1,13):
    print(f'{month}월 기사를 크롤링합니다.')
    print("**************************************")
    if month < 10:
        for day in tqdm(range(1,32)):   
            # URL에 접속 시도
            try:
                # 1 ~ 9일에 대한 URL 접근
                if day < 10:
                    URL = f'https://media.naver.com/press/{presscode}/newspaper?date=2022{month}0{day}'
                    
                    response = requests.get(URL)
                    print(response)
                # 10일부터에 대한 URL 접근
                else:
                    URL = f'https://media.naver.com/press/{presscode}/newspaper?date=2022{month}0{day}'
                    
                    response = requests.get(URL)
                    
                    print(response)
                # html 파싱 시작
                soup = BeautifulSoup(response.text)
                page = soup.select('div.newspaper_inner')
               
                # 혹시나 1면도 없으면 신문 없는 날
                if len(page) < 1:
                    print(f'{day}일자는 기사가 없습니다 데헷')
                    print('-' * 25)
                    continue

                # 신문 지면 발견하면 크롤링
                else:
                    print(f'{day}일자 기사를 크롤링합니다.')
                    print('-' * 25)

                    # 하루치 기사를 담을 리스트
                    oneday_titles = []

                    # 지면별 헤드라인 탐색
                    for _ in range(len(page)):
                        page_titles = []
                        title = soup.select('div.newspaper_inner')[_]
                        titles = title.find_all('strong')

                        # 헤드라인들 page_title 리스트에 담음
                        for j in titles:
                            page_titles.append(j.text)

                    # 하루치 기사 담아주기(1)
                        oneday_titles.append(page_titles)


                    if day < 10:

                        df.iloc[len(df)]=[to_crawl_press_this_time,oneday_titles,date]
                    else:

                        df.iloc[len(df)]=[to_crawl_press_this_time,oneday_titles,date]



        # URL 접속 안되면 신문 없는 날    
            except:
                print(f'{day}일자는 기사가 없습니다 데헷!')
                print('-' * 25)
                continue
    else:
        for day in tqdm(range(1,32)):
            # URL에 접속 시도
            try:
                # 1 ~ 9일에 대한 URL 접근
                if day < 10:
                    date=f'2021{month}0{day}'

                    response=requests.get(url=f'https://media.naver.com/press/{presscode}/newspaper',headers=request_headers
                                         ,params={'date':date})
                # 10일부터에 대한 URL 접근
                else:
                    date=f'2021{month}{day}'
                    response=requests.get(url=f'https://media.naver.com/press/{presscode}/newspaper',headers=request_headers
                                         ,params={'date':date})
                print(response)
                # html 파싱 시작
                soup = BeautifulSoup(response.text)
                page = soup.select('div.newspaper_inner')

                # 혹시나 1면도 없으면 신문 없는 날
                if len(page) < 1:
                    print(f'{day}일자는 기사가 없습니다 데헷')
                    print('-' * 25)
                    continue

                # 신문 지면 발견하면 크롤링
                else:
                    print(f'{day}일자 기사를 크롤링합니다.')
                    print('-' * 25)

                    # 하루치 기사를 담을 리스트
                    oneday_titles = []

                    # 지면별 헤드라인 탐색
                    for _ in range(len(page)):
                        page_titles = []
                        title = soup.select('div.newspaper_inner')[_]
                        titles = title.find_all('strong')

                        # 헤드라인들 page_title 리스트에 담음
                        for j in titles:
                            page_titles.append(j.text)

                    # 하루치 기사 담아주기(1)
                        oneday_titles.append(page_titles)


                    if day < 10:

                        df.iloc[len(df)]=[to_crawl_press_this_time,oneday_titles,date]
                    else:

                        df.iloc[len(df)]=[to_crawl_press_this_time,oneday_titles,date]



        # URL 접속 안되면 신문 없는 날    
            except:
                print(f'{day}일자는 기사가 없습니다 데헷!')
                print('-' * 25)
                continue
```
