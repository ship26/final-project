```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd
import numpy as np
```

```python
response=requests.get('https://media.naver.com/press/015/newspaper',
             params={'date':'20210402'}
            )
```

```python
soup=BeautifulSoup(response.text)
soup
```

```python

면=soup.select('div.newspaper_inner')
```

```python
타이틀제목들=[]
for i in range(len(면)):
    i면타이틀=[]
    타이틀제목=soup.select('div.newspaper_inner')[i]
    타이틀제목s=타이틀제목.find_all('strong')
    for j in 타이틀제목s:
        i면타이틀.append(j.text)
    타이틀제목들.append(i면타이틀)
타이틀제목들
```
