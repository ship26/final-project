# emotion 한글화

```python
def ko(x):
    if x==0.0:
        x='부정'
    elif x==1.0:
        x='긍정'
    else:
        x='중립'
    return x

df['emotion']=df['emotion'].apply(ko)
len(df)
```

```python
df.reset_index()
df.to_csv('traain.csv',encoding='utf-8-sig')

```

```python
train=df.sample(frac=0.8,random_state=2022)
len(train)
train
```

```python
test=df.drop(train.index)
len(test)
test=test.reset_index(drop=True)
test
train=train.reset_index(drop=True)
train
print(len(train),len(test))
```

```python
test.to_csv('final_test_data.csv',encoding='utf-8-sig')
train.to_csv('final_train_data.csv',encoding='utf-8-sig')
```
