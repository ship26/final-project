```python
df=pd.read_csv('data/common_percent.csv',encoding='utf-8-sig',index_col=0)
df
```

```python
df.press.nunique()
```

```python
한국경제=df.loc[df['press']=='한국경제']
한국경제['부정']=round(한국경제['뉴스개수']*한국경제['0_percent'])
한국경제['긍정']=round(한국경제['뉴스개수']*한국경제['1_percent'])
한국경제['중립']=round(한국경제['뉴스개수']*한국경제['2_percent'])
한국경제
```

```python
경향신문=df.loc[df['press']=='경향신문']
경향신문['부정']=round(경향신문['뉴스개수']*경향신문['0_percent'])
경향신문['긍정']=round(경향신문['뉴스개수']*경향신문['1_percent'])
경향신문['중립']=round(경향신문['뉴스개수']*경향신문['2_percent'])
경향신문
```

```python
merge1=pd.merge(한국경제,경향신문,left_on='date'  ,right_on='date',how='outer'   )
merge1.fillna(0,inplace=True)
merge1['뉴스개수']=merge1['뉴스개수_x']+merge1['뉴스개수_y']
merge1['긍정']=merge1['긍정_x']+merge1['긍정_y']
merge1['중립']=merge1['중립_x']+merge1['중립_y']
merge1['부정']=merge1['부정_x']+merge1['부정_y']
merge1['0_percent']=merge1['부정']/merge1['뉴스개수']
merge1['1_percent']=merge1['긍정']/merge1['뉴스개수']
merge1['2_percent']=merge1['중립']/merge1['뉴스개수']
merge1['press']='일반언론'
merge1=merge1[['date','press','0_percent','1_percent','2_percent','뉴스개수']]
merge1
```
