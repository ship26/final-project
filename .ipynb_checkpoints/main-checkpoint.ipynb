```python
# requirements
import schedule
from datetime import datetime, timedelta 
import import_ipynb
import pandas as pd
import numpy as np
import tqdm
from data_crawl_module import Crawler  #일반뉴스 및 금융뉴스 크롤링
from SC_module import SentimentClassifier as sc  #모델을 통한 감성분석 
from Visualize_module import Emotional_info as emt #시각화 클래스
from final_crawling import Crawler_Stock   #종목별 크롤링
import Keybert_Module as km
#from Keybert_Module import dfchange # 데이터프레임 처리 클래스
#from Keybert_Module import Keybert # 키버트 클래스

today = str(datetime.now())[:10]  ## 오늘 날짜 정의
```

## 전처리 함수

```python
import re

def preprocessing(x):
    
    x = re.sub(r'\[[^)]*\]', '',x)
    x = re.sub(r'\([^)]*\)', ' ',x)
    return x
    
```

# 하루 뉴스 크롤링 & 태깅 & 시각화 예시

```python
# 크롤링과 파일 병합

c = Crawler()
c.crawling_today()
c.crawling_today_finance()
c.merge()
c.merge_finance()

today_df = pd.read_csv(f'data/common/daily/{today}.csv', encoding='utf-8-sig', index_col=0)
today_df_f=pd.read_csv(f'data/finance/daily/{today}.csv', encoding='utf-8-sig', index_col=0)
```

## 네이버금융뉴스와 일반뉴스 합침 ->일반 뉴스 통합

```python

today_df=pd.concat([today_df,today_df_f])
today_df['title']=today_df['title'].apply(preprocessing)
today_df
```

```python
# 감성정보 태깅

sc_today = sc()
labeled_today_df = sc_today.labeling(today_df)
```

```python
# 시각화
labeled_today_df_info = emt(labeled_today_df)
```

```python
labeled_today_df_info.DataFrame
```

## 1-1)전체 분석 이미지, 키워드이미지, 온도계 이미지, 워드 클라우드

```python
import pickle


daily_news={}  #텍스트 를 담을 통

common_ratio=labeled_today_df_info.ratio()  #  원 이미지 저장
common_bar=labeled_today_df_info.bar()  #막대 이미지 저장

from PIL import Image
low = Image.open("low.png")
neutral = Image.open("neutral.png")
high = Image.open("high.png")

#온도 측정
dfchanger = km.dfchange(labeled_today_df_info.DataFrame)
temp = dfchanger.get_temp()

print(f"오늘의 일반뉴스 온도는 {temp} 입니다.")

#온도계 이미지 저장
if temp > 42:
    high.save("daily_news/all/thermometer.PNG",'PNG')
elif temp <=30:
    low.save("daily_news/all/thermometer.PNG",'PNG')
else:
    neutral.save("daily_news/all/thermometer.PNG",'PNG')
    
#워드클라우드 이미지 저장
keybertmodel = km.Keybert(labeled_today_df_info.DataFrame)
wc=keybertmodel.get_wordcloud_c()
wc.to_file('daily_news/all/word_cloud.PNG')

#일반뉴스 키워드 저장
keyword_c=keybertmodel.get_Keyword_c(Mode='MMR',n=3,nrc=10,diversity=0.2)
daily_news['all']=keyword_c
```

## 1-2) 긍정 분석 이미지, 키워드 이미지 + 관련 뉴스링크

```python
positive_labeld_today_df=dfchanger.get_1()
positive_plt=km.dfchange(positive_labeld_today_df)

positive_plt.ratio(file='positive',key=None)  #  원 이미지 저장
positive_plt.bar(file='positive',key=None)  #막대 이미지 저장


    
#긍정워드클라우드 이미지 저장
keybertmodel = km.Keybert(positive_labeld_today_df)
wc=keybertmodel.get_wordcloud_c()
wc.to_file('daily_news/positive/word_cloud.PNG')

#긍정뉴스 키워드 저장
keyword_c=keybertmodel.get_Keyword_c(Mode='MMR',n=3,nrc=10,diversity=0.2)

#긍정뉴스 기사 링크 저장
temp_p = km.Keybert(positive_labeld_today_df)
p_link = temp_p.get_link(emotion='긍정')

p_link = p_link[0][:5]
p_link = p_link.values


positive_list=[]
positive_list.extend(keyword_c)
positive_list.extend(p_link)
print(positive_list)
daily_news['positive']=positive_list  #데이터프레임에 저장
print(daily_news)
```

## 1-3) 부정 분석 이미지, 키워드 이미지 + 관련 뉴스링크

```python
negative_labeld_today_df=dfchanger.get_0()
negative_plt=km.dfchange(negative_labeld_today_df)

negative_plt.ratio(file='negative',key=None)  #  원 이미지 저장
negative_plt.bar(file='negative',key=None)  #막대 이미지 저장


    
#부정워드클라우드 이미지 저장
keybertmodel = km.Keybert(negative_labeld_today_df)
wc=keybertmodel.get_wordcloud_c()
wc.to_file('daily_news/negative/word_cloud.PNG')

#부정뉴스 키워드 저장
keyword_c=keybertmodel.get_Keyword_c(Mode='MMR',n=3,nrc=10,diversity=0.2)
print(keyword_c)
#부정뉴스 기사 링크 저장
temp_n = km.Keybert(negative_labeld_today_df)
n_link = temp_n.get_link(emotion='부정')

n_link = n_link[0][:5]
n_link = n_link.values


negative_list=[]
negative_list.extend(keyword_c)
negative_list.extend(n_link)
print(negative_list)
daily_news['negative']=negative_list  #데이터프레임에 저장
print(daily_news)
## 피클로 문자 파일 저장

# save
with open('daily_news.pickle', 'wb') as f:
    pickle.dump(daily_news, f, pickle.HIGHEST_PROTOCOL)
```

2. 종목별, 종목명 또는 티커 입력받아 해당 기업 선택 (72개사)


```python
df_sector ={'Food':{},'Clothing':{},'Chemical':{},'Medicine':{},
        'Non_Metal':{},'Metal':{},'Machine':{},'Electronic':{},
        'Construction':{},'Transport':{},'Distribution':{},
        'Power':{},'Tele':{},'Finance':{},'Brokerage':{},
        'Insurer':{},'Service':{},'Manufacturer':{}}
```

```python

CS = Crawler_Stock()   #종목별 크롤링 클래스 생성자

Stock_list = CS.Stock_list  #종목별 기업별 리스트

stock_L = []
for i in Stock_list.values():
    stock_L.extend(i)
    
```

2-1) 워드클라우드 이미지, 키워드 텍스트 + 관련 뉴스링크

```python

for st in stock_L:

    ticker = st #티커입력받음

    CS.crawl_news_data(ticker,1)  #티커의 1페이지까지 크롤링

    today_df_sector = pd.read_csv(f'{ticker}_test.csv',encoding='utf-8-sig',index_col=0)


    for key, value in Stock_list.items() :
        if ticker in value:
            sector_key = key   #티커에 맞는 키 저장


    sc_sector = sc()   #감성태깅
    labeled_today_sector_df = sc_sector.labeling(today_df_sector)  #태깅된 데이터프레임


    sector_plt=km.dfchange(labeled_today_sector_df)
    sector_plt.ratio(file=ticker,key=sector_key)  #  원 이미지 저장
    sector_plt.bar(file=ticker,key=sector_key)  #막대 이미지 저장

    #온도 측정
    dfchanger_sector = km.dfchange(labeled_today_sector_df)
    temp_sector = dfchanger_sector.get_temp()   # 온도 수치
    temp_sector_char=dfchanger_sector.get_this()  #부정이냐 긍정이냐의 값



    if temp_sector_char=='긍정':
        sector_labeld_today_df=sector_plt.get_1()
        sector_plt=km.dfchange(sector_labeld_today_df)
    elif temp_sector_char=='부정':
        sector_labeld_today_df=sector_plt.get_0()
        sector_plt=km.dfchange(sector_labeld_today_df)
    else :
        sector_labeld_today_df=sector_plt.get_2()
        sector_plt=km.dfchange(sector_labeld_today_df)

    print(f"오늘의{ticker} 일반뉴스 온도는 {temp_sector} 입니다.")

    #온도계 이미지 저장
    if temp_sector > 42:
        high.save(f"sector/{sector_key}/{ticker}/thermometer.PNG",'PNG')
    elif temp_sector <=30:
        low.save(f"sector/{sector_key}/{ticker}/thermometer.PNG",'PNG')
    else:
        neutral.save(f"sector/{sector_key}/{ticker}/thermometer.PNG",'PNG')


    
    print(f'오늘의 감성수치는 {temp_sector_char} 입니다.')


    
    #워드클라우드 이미지 저장
    keybertmodel_sector = km.Keybert(sector_labeld_today_df)
    wc_sector=keybertmodel_sector.get_wordcloud_c()
    wc_sector.to_file(f'sector/{sector_key}/{ticker}/word_cloud.PNG')

    #일반뉴스 키워드 저장
    keyword_sector=keybertmodel_sector.get_Keyword_c(Mode='MMR',n=3,nrc=10,diversity=0.2)
    sector_list=[]
    sector_list.extend(keyword_sector)

    #링크 저장
    sector_link = keybertmodel_sector.get_link(emotion=temp_sector_char)
    sector_link = sector_link[0][:5]
    sector_link = sector_link.values
    sector_list.extend(sector_link)

    #데이터 저장 및 피클저장


    if sector_key in Stock_list.keys():
        df_sector[sector_key][ticker]=sector_list


# save
with open('df_sector_A.pickle', 'wb') as f:
    pickle.dump(df_sector, f, pickle.HIGHEST_PROTOCOL)
```
